```python
import json
import csv
import urllib.request
import pandas as pd
```

```python
def get_data(latitude,longitude,startdate,enddate):
    url = "http://aura.astralanguage.com/iterismodels/iteris/ten_year_agweather/?Latitude=" + latitude +"&Longitude="+ longitude + "&StartDate="+startdate+"&EndDate="+enddate+"&Key=TFDMHIWS8BH5GP9H6LB041A345OQ7KW"
    response = urllib.request.urlopen(url)
    raw_json = response.read().decode("utf-8")
    data = json.loads(raw_json)
    return data
```

```python
latitude =['52.034454','51.825986','51.823400','51.141100']
longtitude = ['0.349323','0.066960','-1.290500','-0.484000']
startdate = ['2010-01-01','2011-01-01','2012-01-01','2013-01-01','2014-01-01','2015-01-01','2016-01-01','2017-01-01','2018-01-01','2019-01-01','2020-01-01']
enddate = ['2011-01-01','2012-01-01','2013-01-01','2014-01-01','2015-01-01','2016-01-01','2017-01-01','2018-01-01','2019-01-01','2020-01-01','2021-01-01']
```

```python
oldleyfield =dict()
Oldleyfield =dict()
for i in range(11):
    oldleyfield[i] = get_data(latitude[0],longtitude[0],startdate[i],enddate[i])
    oldleyfield[i] =oldleyfield[i]['52.034454,0.349323']
    Oldleyfield = {**Oldleyfield, **oldleyfield[i]}
```

```python
df = pd.DataFrame(Oldleyfield)
df = df.T.drop(['descriptors'], axis=1)
df.head()
```

```python
df.to_csv('North_east.csv')
```

```python
regex = r"([-+]?\d*\.*\d+|\d+)"

def clean_data(data):
    data['Datetime'] = pd.to_datetime(data['Unnamed: 0'])
    data = data.set_index(['Datetime'])
    for col in data.columns:
        data[col]=data[col].str.extract(regex, expand=False)
    data = data.drop(['Unnamed: 0'], axis=1)    
    data =data.dropna(how='all',axis = 1)
    return data

Oldleyfield = clean_data(data)

Oldleyfield.to_csv('North_east.csv')
```
